```python
from os import listdir
from numpy import asarray
from numpy import save
from tensorflow.keras.utils import load_img, img_to_array
folder = '/content/drive/MyDrive/Colab Notebooks/DATA/VN dishes_10/'

photos, labels = list(), list()
for file in listdir(folder):
  if file.startswith('phalau'):
    output = 0
    photo = load_img(folder + file,target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('myxao'):
    output = 1
    photo = load_img(folder + "myxao" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('hutieu'):
    output = 2
    photo = load_img(folder + "hutieu" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('goicuon'):
    output = 3
    photo = load_img(folder + "goicuon" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('comtam'):
    output = 4
    photo = load_img(folder + "comtam" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('comga'):
    output = 5
    photo = load_img(folder + "comga" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('banhmi'):
    output = 6
    photo = load_img(folder + "banhmi" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('bundaumamtom'):
    output = 7
    photo = load_img(folder + "bundaumamtom" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

  if file.startswith('bunbo'):
    output = 8
    photo = load_img(folder + "bunbo" + str(i+1) + ".jpg",target_size=(200,200))
    photo = img_to_array(photo)
    photos.append(photo)
    labels.append(output)

photos = asarray(photos)
labels = asarray(labels)
print(photos.shape,labels.shape)

save('Food_photos.npy', photos)
save('Food_labels.npy', labels)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, BatchNormalization, Dropout, Conv2D, MaxPooling2D, Flatten # BatchNormalization: để dùng GPU ko lỗi

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(200, 200, 3)))
model.add(MaxPooling2D((2, 2), padding = 'same'))

model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
model.add(MaxPooling2D((2, 2), padding = 'same'))

model.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
model.add(MaxPooling2D((2, 2), padding = 'same'))
model.add(Flatten())

model.add(Dense(256, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(10, activation='softmax'))

# Translate
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
 history = model.fit(photos, labels, epochs=100, verbose=1)
```

```python
model.evaluate(photos,labels)
```
